```python
import ikpy.chain
import numpy as np
import ikpy.utils.plot as plot_utils
import matplotlib.pyplot as plt
my_chain = ikpy.chain.Chain.from_urdf_file("spot_arm.urdf")
my_chain.active_links_mask = [False,True,True,True,True,True,True,False]
print(my_chain)

```

```python
zeros = my_chain.forward_kinematics([0]*8)
print(zeros)
```

```python
fig, ax = plot_utils.init_3d_figure()
my_chain.plot([0]*8,ax)
```

```python
pos = [-0.877,0.823505,0.0185]
rot = [0,0,0]
zero_solve = my_chain.inverse_kinematics(target_position=pos,target_orientation=rot,orientation_mode="all")
print(zero_solve)
```

```python
my_chain.plot(zero_solve,ax)
```
